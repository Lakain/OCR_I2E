```python
POPPLER_PATH = os.getenv('POPPLER_PATH')
```

```python
def pdf_to_images(pdf_path):
    return convert_from_path(pdf_path, poppler_path=POPPLER_PATH)
```

```python
def extract_tabular_data_from_images(images):
    all_tables = []
    for image in images:
        # Use Tesseract to extract data including boxes, confidences, and line numbers
        ocr_data = image_to_data(image, output_type='data.frame')
        print(ocr_data)
        ocr_data = ocr_data.dropna().reset_index(drop=True)

        # Filter non-zero confidence rows
        ocr_data = ocr_data[ocr_data.conf != 0]

        # Assuming a very basic table structure where each text piece with the same top position belongs to the same row
        grouped_data = ocr_data.groupby('top')

        table = []
        for _, group in grouped_data:
            row = group.sort_values(by='left').text.tolist()
            table.append(row)

        all_tables.append(table)
    return all_tables
```
